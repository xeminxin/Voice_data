## 원본 오디오와 잡음 생성된 오디오 비교

```python
import soundfile as sf
import matplotlib.pyplot as plt
```

### 원본 오디오 불러오기

```python
# 원본 wav 파일 로드
file_paht = "./sample_data/blues.00001.wav"
data , sr = sf.read(file_paht)

# 샘플 wav 파일 로드
file_path = "./sample_data/sample_noise_add.wav"
noise_data, noise_sr = sf.read(file_path)
```

### 시각화

```python
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(data)
plt.title('Original')
plt.show()
plt.figure(figsize=(20, 5))
plt.subplot(1,2,1)
plt.plot(noise_data)
plt.title("Noise Add")
plt.show()
```
