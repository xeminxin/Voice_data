# 인공신경망 - Relu 함수
입력값이 0보다 작으면 0 출력, 0 이상이면 그대로 출력
- 연산이 빠르고 간단하기 때문에, 계산 속도가 빠름
- 입력값이 0 이하인 경우, gradient가 0이 되어버리므로, 해당 뉴런은 다시 회복할 수 없게 된다.
- 이와 같은 문제를 해결하기 위해, LeakRelu, PReLu와 같은 변형된 ReLu 함수가 제안되기도 한다.

ReLU 함수는 x<0에서는 0으로 수렴하고, x가 0보다 크거나 같은 경우 기울기가 1인 직선 그래프를 가진다.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def relu(x):
    
    return np.maximum(0, x)

x = np.linspace(-5, 5, 1000)
# -5부터 5까지를 100개 구간으로 나눈 값들을 x에 할당한다.
y = relu(x)
print(y)
```

# 인공신경망 - Leaky ReLU 함수
Leaky ReLU 함수는 입력값이 음수인 경우에도 미분 가능. ReLU 함수에서 발생하는 dying ReLU 문제 해결 가능
> dying ReLU : 입력값이 음수일 때는 그 미분값이 0이 되어 역전파 시 가중치 업데이트가 이루어지지 않아 학습이 멈추는 현상

ReLU 함수의 변형으로 음수 범위에서도 0 이외의 값을 가질 수 있도록 설계된 함수
Leaky ReLU 함수는 ReLU 함수와 비슷하지만, 입력값이 음수일 때 0 대신 0에 가까운 매우 작은 값을 반환한다. 이 작은 값을 0.01과 같은 작은 상수값으로 지정할 수도 있다.

```python
def leaky_relu(x, alpha=0.01) :
    return np.where(x >= 0, x, x * alpha)

x = np.linspace(-5, 5, 100)
y = leaky_relu(x)

plt.plot(x, y)
plt.show()
```

Leaky ReLU 함수에서는 추가적인 하이퍼파라미터 alpha가 사용된다. 이 alpha는 음수 영역에서의 함수의 기울기를 결정한다. 기본값으로는 주로 0.01 사용
