# 인공신경망 - 소프트맥스 함수
입력받은 값을 0과 1 사이의 값으로 정규화하며, 모든 클래스에 대한 확률의 합이 1이 되도록 만들어주는 함수
> 3개의 클래스를 분류하는 모델이 있다고 가정했을 때, 이 모델은 출력층에 3개의 뉴런을 가지며, 각 뉴런은 해당 클래스에 속할 확률을 출력한다. 소프트맥스 함수를 거치기 전의 출력값은 일반적으로 logit이라고 부른다. 로짓은 임의의 실수값이 될 수 있으며, 이를 소프트맥스 함수를 통해 0과 1 사이의 값으로 정규화하면 확률값을 얻을 수 있다.

공식에서 K는 벡터 Z 길이이며 softmax 함수를 통해 변환된 벡터는 모든 요소의 합이 1이 된다.
이러한 성질때문에 softmax 함수는 분류 문제에서 각 클래스에 속할 확률을 예측하기 위해 주로 사용된다.

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(z):
    return np.exp(z) / np.sum(np.exp(z))

z = np.linspace(-10, 10, 100)   # -10부터 10까지 100개의 값을 생성하라
y = softmax(z)

plt.plot(z, y)
plt.show()
```

-10부터 10까지의 범위에서 100개의 값을 생성하고, 해당 값을 softmax 함수에 적용하여 반환된 값을 그래프로 시각화 한다.
x축은 입력값 z이고, y축은 해당 z값에 대한 softmax 값이다.
