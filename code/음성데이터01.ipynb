## 스펙트럼 간단한 실습 - 샘플 사운드 이용하여 스펙트럼 이미지 생성

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 음성 파일 로드
filename = "./sample_data/blues.00001.wav"
y, sr = librosa.load(filename)

# 전처리
# 스펙트로그램 생성을 위해 짧은 오디오 클립 추출
duration = 10 # 추출할 클립의 길이(10초)
samples = int(duration * sr)    # 추출할 샘플 수
start = 0    # 추출 시작지점  (0초부터 시작)
y = y[start:start+samples]

# 스펙트로그램 생성
spec = librosa.stft(y)
spec_db = librosa.amplitude_to_db(np.abs(spec), ref=np.max)

# 스펙트로그램 시각화
librosa.display.specshow(spec_db, sr=sr, y_axis='log', x_axis='time')

# 이미지 파일로 저장
plt.savefig('spectrogram.png')
plt.close()
```
